### Цифрование категориальных/бинарных значений (по принципу: 0 - самый встречающийся объект)
Stage_fear: No - 0, Yes - 1


Drained_after_socializing: No - 0, Yes - 1

Personality: Extrovert - 0, Intovert - 1

```python
train = df_train.dropna()

train['Stage_fear'] = (train['Stage_fear'] != "No").astype(int)
train['Drained_after_socializing'] = (train["Drained_after_socializing"] != 'No').astype(int)
train["Personality"] = (train["Personality"] != "Introvert").astype(int)

train.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
X = train.drop(columns = ['Personality'])
y = train['Personality']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
LogisticModel = LogisticRegression()
LogisticModel.fit(X_train, y_train)
```

```python
def accuracy(x,y):
  assert len(x) == len(y), "Длины не совпадают"
  return sum(x == y) / len(x)
```

```python
pred1 = LogisticModel.predict(X_test)
print(f"Accuracy: {accuracy(pred1, y_test)}")
```

```python
from xgboost import XGBClassifier

bst = XGBClassifier(n_estimators=2, max_depth=2,
                    learning_rate=1, objective='binary:logistic')

bst.fit(X_train, y_train)
pred2 = bst.predict(X_test)
```

```python
print(f"Accuracy: {accuracy(pred2,y_test)}")
```

```python
from sklearn.neighbors import KNeighborsClassifier

KNNModel = KNeighborsClassifier()

KNNModel.fit(X_train, y_train)
pred3 = KNNModel.predict(X_test)
```

```python
print(f"Accuracy: {accuracy(pred3, y_test)}")
```

```python
import tqdm
import numpy as np
```

```python
class BlendingEnsemble:

  def __init__(self, models):
    self.models = models


  def fit(self, X, y):

    for i, model in tqdm.tqdm(enumerate(self.models, 1)):
      print(f"Начало обучения модели номер: {i}")
      model.fit(X, y)
      print(f"Конец обучения модели номер: {i}")


  def predict(self, X):

    pred = np.zeros(shape = (X.shape[0],))

    for model in self.models:
      pred += model.predict(X)

    return np.array([1 if i else 0 for i in (pred / 3 >0.5)])
```

```python
model1 = LogisticRegression()
model2 = KNeighborsClassifier()
model3 = XGBClassifier(n_estimators=2, max_depth=2,
                       learning_rate=1, objective='binary:logistic')

models = [model1, model2, model3]
```

```python
BLEnsemble = BlendingEnsemble(models)
```

```python
BLEnsemble.fit(X_train, y_train)
```

```python
pred4 = BLEnsemble.predict(X_test)
print(f"Accuracy: {accuracy(pred4, y_test)}")
```

```python
import optuna
from sklearn.model_selection import cross_val_score
```

```python
from catboost import CatBoostClassifier
```

```python
def objective(trial):
    params = {
        'n_estimators': trial.suggest_int('n_estimators', 50, 300),
        'max_depth': trial.suggest_categorical('max_depth', [2, 4, 6]),
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.1),
        'l2_leaf_reg': trial.suggest_categorical('l2_leaf_reg', [1, 3, 5])
    }

    model = CatBoostClassifier(**params, random_state=42, verbose=False)
    scores = cross_val_score(model, X, y, cv=5, scoring='accuracy')

    return scores.mean()
```

```python
study = optuna.create_study(direction='maximize')

study.optimize(objective, n_trials=50, show_progress_bar=True)

print("Лучшие гиперпараметры:")
for key, value in study.best_params.items():
    print(f"{key}: {value}")

print(f"Лучшая точность: {study.best_value:.4f}")
```

```python
study.best_params
```

```python
from sklearn.model_selection import KFold
from catboost import Pool
```

```python
NFOLDS = 5
folds = KFold(n_splits=NFOLDS, shuffle=True, random_state=42)
CatBoostModels = []
best_params = study.best_params
best_params['n_estimators'] = 500
best_params['verbose'] = 100
best_params['eval_metric'] = 'Accuracy'
```

```python
for fold_, (trn_idx, val_idx) in enumerate(folds.split(X, y)):
    print(f"\n--- Starting Fold {fold_ + 1}/{NFOLDS} ---")

    train_pool = Pool(X.iloc[trn_idx], y.iloc[trn_idx])
    eval_pool = Pool(X.iloc[val_idx], y.iloc[val_idx])

    model = CatBoostClassifier(**best_params)

    model.fit(train_pool,
              eval_set=eval_pool,
              use_best_model=True)

    CatBoostModels.append(model)
```

```python
def PredictCatBoostEnsemble(models, x):
  pred = np.ones(len(x))

  for model in models:
    pred += model.predict(x)
  return np.array([1 if i >=3 else 0 for i in pred])
```
